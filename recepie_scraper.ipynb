```python
def get_links_from_page(url,filter = 'recipe'): 
    return list(set([l['href'] for l in scrape_me(url).links() if f'.com/{filter}/' in l['href']]))
```

```python
links = get_links_from_page('https://www.allrecipes.com/','recipes');len(links)
```

```python
reslinks = [ll for l in links for ll in get_links_from_page(l,'recipe')];len(reslinks)
```

```python
scrape_me(reslinks[0]).nutrients()
```

```python
def get_recepie(url):
    scraper = scrape_me(url)
    return (scraper.title(),
            scraper.total_time(),
            scraper.yields(),
            scraper.ingredients(),
            scraper.instructions(),
            scraper.image(),
            scraper.host(),
            # scraper.links(),
            scraper.nutrients())
```
