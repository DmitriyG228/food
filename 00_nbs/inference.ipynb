```python
l = 400
```

```python
total = 1
```

```python
total
```

```python
while total >0:
    try:
        l-=1

        query = f"""select id,text
                     from food.{table}
                    where clip is Null """

        total = engine.execute(f'select count(*) from ({query}) a').one()[0]
        pd_iter = cpd.read_sql_query(query, engine, chunksize=bs, index_col='id')
        for inp in tqdm(pd_iter, desc="clip food inference", total=total // bs):  
            text = inp['text'].iloc[0]
            clip = get_text_clip(text[:l])
            stmt = (update(FoodsP).
                    values(clip=[c.item() for c in clip]).
                    where(FoodsP.id == int(inp.index[0])))
            engine.execute(stmt)
    except:
        pass
```
