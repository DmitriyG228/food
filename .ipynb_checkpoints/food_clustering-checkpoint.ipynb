```python
series2tensor = lambda series:torch.tensor([np.array(c) for c in series.values])
```

```python
def norm(v): 
    return v/torch.linalg.norm(v, dim=-1, keepdim=True)
```

```python
foods = read_sql('foods')

# foods = foods.drop(columns = ['clip'])
foods = foods
foods = foods.dropna()
```

```python
foods = foods.replace(0,0.1)
```

```python
food_clips = series2tensor(foods['clip'])
```

```python
nutrition_tensor = norm(series2tensor(foods[['energy','protein','carb','fat']]))
```

```python
food_clips = norm(torch.concat([food_clips,nutrition_tensor],dim=1))
```

```python
food_clips = norm(food_clips)
```

```python
food_clips, _, _= torch.pca_lowrank(food_clips, q=8, center=True, niter=50)
```
